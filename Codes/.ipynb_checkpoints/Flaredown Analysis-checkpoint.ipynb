```python
##select Crohn Disease

dfCrohn=df[df.trackable_name.isin (['Crohns','Crohn‚Äôs Disease','Crohn','Crohns disease','Spondylitis , crohns disease,copd',
                                    'Crohns\'s disease'])]
dfCrohn.shape
```

```python
#Select treatment for Crohn
df2=df[['user_id', 'checkin_date','trackable_type','trackable_name','trackable_value']]
df2food=df2[df2.trackable_type=='Treatment']
df2food.shape
```

```python
treatment=pd.merge(dfCrohn,df2food, on=['user_id','checkin_date'])
treatment1=treatment.copy()
#dates to standard datetime
import time
from datetime import datetime
treatment['checkin_date']=pd.to_datetime(df.checkin_date,format = '%m/%d/%y')

print(treatment.shape)

```

```python
treatment.head()
```

```python
#find dates with diffrent format
listdate=[]
for i in range(len(treatment)):
    if pd.isnull(treatment.iloc[i,4]):
        listdate.append(i)

listdate
```

```python
for i in listdate:
    treatment.iloc[i,4]=pd.to_datetime(treatment1.iloc[i,4],format = '%m/%d/%y')
```

```python
treatment['medicinedose']=np.zeros(treatment.shape[0])

for i in range(len(treatment)):
    if str(treatment.iloc[i,-2]).find('one half')!=-1:
        treatment.iloc[i,-1]=1.5
    elif str(treatment.iloc[i,-2]).find('half')!=-1:
        treatment.iloc[i,-1]=0.5
    elif str(treatment.iloc[i,-2]).find('x')==-1:
        treatment.iloc[i,-1]=str(treatment.iloc[i,-2]).split(' ')[0]
    elif str(treatment.iloc[i,-2]).find('x')!=-1:
        treatment.iloc[i,-1]=str(treatment.iloc[i,-2]).replace('mg','')
        
 

        





    
```

```python
import numpy
for i in range(len(treatment)):
    if str(treatment.iloc[i,-1]).find('x')!=-1:
        treatment.iloc[i,-1]=str(treatment.iloc[i,-1]).replace('puff','')
        
        listmedicine=str(treatment.iloc[i,-1]).split('x')
        #print(listmedicine)
        treatment.iloc[i,-1]=float(listmedicine[0])*float(listmedicine[1])
        
   
for i in range(len(treatment)):
    if str(treatment.iloc[i,-1]).find('mg')!=-1:
        treatment.iloc[i,-1]=str(treatment.iloc[i,-1]).replace('mg','')
    elif str(treatment.iloc[i,-1]).find('g')!=-1:
        treatment.iloc[i,-1]=str(treatment.iloc[i,-1]).replace('g','')
    elif str(treatment.iloc[i,-1]).find('/day')!=-1:
        treatment.iloc[i,-1]=str(treatment.iloc[i,-1]).replace('/day','')
    

treatment.drop([41],inplace=True)     
```

```python
grouped=treatment.groupby('user_id')
recordfirst=grouped.agg(lambda x: x.iloc[0])['medicinedose']
recordlast=grouped.agg(lambda x: x.iloc[-1])['medicinedose']

print(recordlast)
selectedid=[]
for i in range(len(recordfirst)):
    if recordfirst[i]>recordlast[1]:
        selectedid.append(i)
print('Userindex with disease improvement with index',selectedid)

selecteduser=[]
for j in selectedid:
    #print(recordlast.index[j])
    selecteduser.append(recordlast.index[j])

print('User_id with disease improvement with id',selecteduser)
```

## Find users common food and treatment

```python
# Select rows with last_name equal to some values, all columns
#data.loc[data['first_name'].isin(['France', 'Tyisha', 'Eric'])]  
treamentselected=treatment.loc[treatment['user_id'].isin(selecteduser)]
print(treamentselected.shape)
treamentselected.groupby('trackable_name_y').count().iloc[:,0].sort_values(ascending=False)
```
